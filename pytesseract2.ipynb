```python
import os
from PIL import Image
import pytesseract

folder_path = r"E:\convert pdf\resize screens"
output_folder_path = r'E:\convert pdf\screens to text'

# Loop through each image in the specified folder
for image_name in os.listdir(folder_path):
    image_path = os.path.join(folder_path, image_name)

    # Open the image using PIL
    image = Image.open(image_path)

    # Use PyTesseract to extract text with Arabic language
    text = pytesseract.image_to_string(image, lang='ara', config='--oem 1')

    # Specify the output file path
    output_file_path = os.path.join(output_folder_path, image_name + '.txt')

    # Save the extracted text to the output file
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        output_file.write(text)

# Print "Done" after processing all images
print("Done")
```

```python
def process_arabic_sentence(sentence, all_arabic_words):
    # Split the sentence into words
    words = sentence.split()

    # Process each word
    modified_words = []
    for word in words:
        # Check if the word ends with "ء" and is not in the list
        if word.endswith("ء") and word not in all_arabic_words:
            # Remove "ء" from the word
            modified_word = word[:-1]
            modified_words.append(modified_word)
        else:
            modified_words.append(word)

    # Join the modified words back into a sentence
    modified_sentence = " ".join(modified_words)

    return modified_sentence

# Sample list of Arabic words
# arabic_words = ["بدء", "باء", "جزء", "جاء", "داء", "دفء", "سوء", "ساء", "شاء", "شيء", "ضوء", "عبء", "فاء", "قاء", "كفء", "ملء", "ماء", "ناء", "هاء", "ياء"]

# Sample Arabic sentence
arabic_sentence = "حشود البشرء الهائمة على وجهها على إيقاع تسكع صيفى لا مبال قوق الأرصفة غير المستوية لمدينة القاهرة العتيقة. بدت وكأنها قد تكيفت. بسكينة بل وبشىء من السخرية اللاذعة. مع تدهور البيئة المستمر الذى لا رجعة فيه. وربما تحدثا انفسنا بأن مجمل هؤلاء الجسورين. المتنزهين تحت الحمم الجارفة لشمس منصهرة. متواطئثون بتسامح, فى تجوالهم الذى لا يكل مع العدو الخفى المقوض لقواعد وأساسات عاصمة كانت متارة فى الماضى. تلك الجماهير التى لا تؤثر فيها المأساة ولا حتى الحزن اشبه يسيل بشرى جارف يحمل معه عينات متنوعة من البشر أصابتها البطالة بالسكينة: عمال عاطلون. حرفيون بلا زبائن. مفكرون خاب أملهم فى بلوغ قمة المجد؛ موظفون إداريون مطرودون من مكاتبهم لعجز فى عدد المقاعد. خريجو جامعات رازحون تحت وطأة علمهم العقيم الذى لم يت ثماره. وأخيرًا. الهازثون الأزليون. هؤلاء الفلاسفة المحيون للظلام والعتمة ولهدوثهم؛ الذين يرون أن هذا التدهور المشهود لمدينتهم قد صمم خصيصنًا لشحذ حاستهم الثقدية. وقد"
# Process the Arabic sentence
modified_sentence = process_arabic_sentence(arabic_sentence, all_arabic_words)

# Print the modified sentence
print("Original Sentence:", arabic_sentence)
print("Modified Sentence:", modified_sentence)
```

```python
import soundfile as sf
import numpy as np
```

```python
import os
```

```python
def merge_audio_files(folder_path, output_file_path):
    # Get all audio files in the specified folder
    audio_files = [file for file in os.listdir(folder_path) if file.lower().endswith(('.wav', '.mp3', '.ogg'))]

    # Load each audio file
    audio_data = []
    for file in audio_files:
        file_path = os.path.join(folder_path, file)
        data, samplerate = sf.read(file_path)
        audio_data.append(data)

    # Concatenate the audio data
    concatenated_audio = np.concatenate(audio_data)

    # Write the concatenated audio to a file
    sf.write(output_file_path, concatenated_audio, samplerate)

    print("Audio files concatenated successfully.")
```
